```python
import pandas as pd
import geopandas as gpd
import os
os.listdir()
```

```python
bstop = pd.read_csv('국토교통부_전국 버스정류장 위치정보_20231016.csv', encoding='cp949')
bstop.head()
```

```python
gb_bstop = bstop[bstop['도시명'].apply(lambda x: '경상북도' in x)]
gb_bstop['도시명'].unique()
```

```python
bstop_gdf = gpd.GeoDataFrame(data=gb_bstop,
                 geometry=gpd.points_from_xy(gb_bstop['경도'], gb_bstop['위도']),
                 crs='epsg:4326')
```

```python
bstop_gdf.columns = ['STOP_NO', 'STOP_NM', 'LONG', 'LAT', 'DATE', 'MOBILE', 'CITY_CODE', 'CITY_NAME', 'ADM_CITY', 'geometry']
bstop_gdf
```

```python
#  shp 파일로 저장할 때는 컬럼명이 전부 영어로 되어있어야 함
bstop_gdf.to_file('gb_bustop.shp', encoding='utf-8')
```
